Bài 5: Viết chương trình giải phương trình dạng bậc 2 (ax^2 + bx + c = 0)

```python
# Cách 1:
a, b, c = map(float, input().split())
dau1 = "-" if b < 0 else "+"
dau2 = "-" if c < 0 else "+"
print(f"Phương trình có dạng: {a:g}x^2 {dau1} {abs(b):g}x {dau2} {abs(c):g} = 0")
delta =  b ** 2 - (4 *a * c)
if delta < 0:
    print("Phương trình vô nghiệm")
elif delta == 0:
    print("Phương trình có nghiệm kép:" , f"{-b / (2 *a):.1f}")
elif delta > 0:
    x1 = (-b + (delta ** 0.5)) / (2 * a)
    x2 = (-b - (delta ** 0.5)) / (2 * a)
    print("Phương trình có hai nghiệm phân biệt:", "x1 =", f"{x1:.1f}",", " "x2 =", f"{x2:.1f}")
else:
    print("Đầu vào không hợp lệ")
```

```python
# Cách 2:
import math
def giai_pt():
    print("Giải phương trình bậc hai ax^2 + bx + c = 0")
    a, b, c = map(float, input("Vui lòng nhập lần lượt: a, b, c: ").split())

    if a == 0:
        if b == 0:
            if c == 0:
                print("Phường trình có vô số nghiệm")
            else:
                print("Phương trình vô nghiệm")
        else:
            print(f"Phương trình có 1 nghiệm duy nhất: x = {-c/b}")
        return
    delta = (b ** 2) - (4 * a * c)

    if delta > 0:
        x1 = (-b + math.sqrt(delta)) / (2 * a)
        x2 = (-b - math.sqrt(delta)) / (2 * a)
        print(f"Phương trình có 2 nghiệm phân biệt:\nx1 = {x1}\nx2 = {x2}")
    elif delta == 0:
        x = -b / (2 * a)
        print(f"Phương trình có nghiệm kép: x1 = x2 = {x}")
    else:
        print("Phương trình vô nghiệm")
              
giai_pt()
```

Bài 8: Giải hệ phương trình tuyến tính

```python
# Cách 1:
a, b, c, d, m, n = map(float, input().split())
detA = (a * d) - (b * c)
detA1 = (m * d) - (b * n)
detA2 = (a * n) - (c * m)
if detA == 0 and detA1 != 0 and detA2 != 0:
    print("He phuong trinh vo nghiem")
elif detA == 0 and detA1 == 0 and detA2 == 0:
    print("He phuong trinh co vo so nghiem")
else:
    x1 = detA1 / detA
    x2 = detA2 / detA
    print("Nghiem cua he phuong trinh la:", f"x1 = {x1:.1f}", f", x2 = {x2:.1f}")
```

```python
# Cách 2:
import numpy as np
def giai_he():
    print("Hệ phương trình có dạng:")
    print("ax + by = m")
    print("cx + dy = n")

    a, b, c, d, m, n = map(float, input("Vui lòng nhập theo thứ tự: a, b, c, d, m, n").split())

    A = np.array([
        [a, b],
        [c, d]
    ])

    B = np.array([m, n])

    try:
        nghiem = np.linalg.solve(A, B)
        print(f"Nghiệm của hệ phương trình là: x1 = {nghiem[0]}, x2 = {nghiem[1]}")
    except np.linalg.LinAlgError:
        print("Hệ vô nghiệm hoặc vô số nghiệm")

giai_he()
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một điểm A(x,y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
# Cách 1:
a, b, r = map(int, input().split())
x, y = map(int, input().split())
print("Điểm A có tọa độ:", f"({x},{y})")
if ((x-a) ** 2) + ((y - b) ** 2) < (r ** 2):
    print("Điểm A thuộc đường tròn")
elif ((x-a) ** 2) + ((y - b) ** 2) == (r ** 2):
    print("Điểm A thuộc đường tròn")
else:
    print("Điểm A không thuộc đường tròn")
```

```python
# Cách 2:
import math
def diem():
    print("Vui lòng nhập thông tin đường tròn")
    a, b, r = map(int, input("Nhập theo thứ tự: hoành độ x, tung độ y, bán kính r:").split())
    if r < 0:
        print("Bán kính không hợp lệ")
        return
    print("Vui lòng nhập tọa độ điểm A")
    x, y = map(int, input("Nhập theo thứ tự: hoành độ x, tung độ y:").split())

    khoang_cach = math.sqrt((x - a) ** 2 + (y - b) ** 2)

    TOL = 1e-9
    print("KẾT QUẢ:")
    if abs(khoang_cach - r) < TOL:
        print(f"Điểm A({x}, {y}) nằm trên đường tròn")
    elif khoang_cach < r:
        print(f"Điểm A({x}, {y}) nằm trong đường tròn")
    else:
        print(f"Điểm A({x}, {y}) nằm ngoài đường tròn")
diem()

```
