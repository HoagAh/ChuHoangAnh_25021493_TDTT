```python
PI = 3.14
chieu_rong = float(input("Chieu rong = "))
chieu_dai = float(input("Chieu dai = "))
dien_tich_trong_cay = (chieu_dai * chieu_rong) - (PI * (chieu_rong / 2) ** 2)
round_dientich = round(dien_tich_trong_cay, 2)
print("Dien tich trong cay =", round_dientich)
```

```python
a, b, c = list(map(int, (input().split())))
if ((a + b > c) and (a + c > b) and (b + c > a)):
    nua_chu_vi = (a + b + c) / 2
    dien_tich = (nua_chu_vi * (nua_chu_vi - a) * (nua_chu_vi - b) * (nua_chu_vi - c)) ** 0.5
    round_lamtron = round(dien_tich, 1)
    print(round_lamtron)
else:
    print("Khong phai 3 canh cua tam giac")
```
